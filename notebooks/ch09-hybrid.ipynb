# Chapter 9: Hybrid Pattern Playground

Experiment with dynamic pattern routing.

```python
from dataclasses import dataclass
from typing import Dict

@dataclass
class TaskCharacteristics:
    complexity: float = 0.5      # 0-1: simple to complex
    uncertainty: float = 0.5     # 0-1: known to unknown
    requires_consensus: float = 0.5  # 0-1: single to multi-perspective
    safety_critical: float = 0.5     # 0-1: low to high stakes
```

```python
class PatternRouter:
    def __init__(self):
        self.weights = {
            "orchestrator": {"complexity": 0.8, "uncertainty": -0.3},
            "swarm": {"uncertainty": 0.8, "complexity": -0.2},
            "council": {"requires_consensus": 0.9},
            "guardian": {"safety_critical": 1.0},
        }
    
    def route(self, chars: TaskCharacteristics) -> str:
        scores = {}
        char_dict = vars(chars)
        for pattern, weights in self.weights.items():
            score = sum(char_dict.get(k, 0) * v for k, v in weights.items())
            scores[pattern] = score
        return max(scores, key=scores.get)
```

```python
router = PatternRouter()

# Test different task profiles
tasks = [
    ("Complex analysis", TaskCharacteristics(complexity=0.9, uncertainty=0.2)),
    ("Open exploration", TaskCharacteristics(uncertainty=0.9, complexity=0.3)),
    ("Group decision", TaskCharacteristics(requires_consensus=0.9)),
    ("High-stakes action", TaskCharacteristics(safety_critical=0.95)),
]

for name, chars in tasks:
    pattern = router.route(chars)
    print(f"{name:20} -> {pattern}")
```
