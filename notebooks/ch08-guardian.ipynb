# Chapter 8: Guardian Pattern Playground

Build and test guardian pipelines.

```python
from dataclasses import dataclass
from enum import Enum

class GuardianDecision(Enum):
    ALLOW = "allow"
    BLOCK = "block"
    MODIFY = "modify"

@dataclass
class GuardianResult:
    decision: GuardianDecision
    message: str
    modified_content: str = None
```

```python
class Guardian:
    def check(self, content: str) -> GuardianResult:
        raise NotImplementedError

class ProfanityGuardian(Guardian):
    def __init__(self, blocked_words: list):
        self.blocked_words = blocked_words
    
    def check(self, content: str) -> GuardianResult:
        for word in self.blocked_words:
            if word.lower() in content.lower():
                return GuardianResult(GuardianDecision.BLOCK, f"Blocked word: {word}")
        return GuardianResult(GuardianDecision.ALLOW, "Content clean")
```

```python
class GuardianPipeline:
    def __init__(self, guardians: list):
        self.guardians = guardians
    
    def process(self, content: str) -> GuardianResult:
        for guardian in self.guardians:
            result = guardian.check(content)
            if result.decision == GuardianDecision.BLOCK:
                return result
        return GuardianResult(GuardianDecision.ALLOW, "All checks passed")
```

```python
# Build a pipeline
pipeline = GuardianPipeline([
    ProfanityGuardian(["badword", "offensive"]),
    # Add more guardians here!
])

# Test it
test_inputs = ["Hello world", "This contains badword", "Normal message"]
for inp in test_inputs:
    result = pipeline.process(inp)
    print(f"{inp[:30]:30} -> {result.decision.value}: {result.message}")
```
