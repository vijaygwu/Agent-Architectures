Experiment with different voting mechanisms.

```python
from collections import Counter
from enum import Enum

class VotingMethod(Enum):
    MAJORITY = "majority"
    UNANIMOUS = "unanimous"
    RANKED_CHOICE = "ranked_choice"
```

```python
def simulate_council_vote(votes: list, method: VotingMethod):
    """Simulate different voting methods - MODIFY THIS!"""
    if method == VotingMethod.MAJORITY:
        counts = Counter(votes)
        winner, count = counts.most_common(1)[0]
        return winner if count > len(votes) / 2 else "No majority"
    elif method == VotingMethod.UNANIMOUS:
        return votes[0] if len(set(votes)) == 1 else "No consensus"
    else:
        return "Implement ranked choice!"
```

```python
# Simulate votes from 5 council members
votes = ["approve", "approve", "reject", "approve", "abstain"]

for method in VotingMethod:
    result = simulate_council_vote(votes, method)
    print(f"{method.value}: {result}")
```
