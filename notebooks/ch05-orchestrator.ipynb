# Chapter 5: Orchestrator Pattern Playground

Experiment with the Orchestrator pattern interactively.

```python
import asyncio
from typing import List, Dict

# Simple mock LLM for experimentation
async def mock_llm(prompt: str) -> str:
    await asyncio.sleep(0.1)  # Simulate latency
    return f"Response to: {prompt[:50]}..."
```

```python
class SimpleOrchestrator:
    def __init__(self, workers: List[str]):
        self.workers = workers
    
    async def plan(self, task: str) -> List[Dict]:
        """Decompose task into subtasks - MODIFY THIS!"""
        return [{"worker": w, "subtask": f"{task} - {w} part"} for w in self.workers]
    
    async def delegate(self, plan: List[Dict]) -> List[str]:
        """Execute subtasks in parallel"""
        tasks = [mock_llm(p["subtask"]) for p in plan]
        return await asyncio.gather(*tasks)
    
    async def aggregate(self, results: List[str]) -> str:
        """Combine results - MODIFY THIS!"""
        return "\n".join(results)
    
    async def execute(self, task: str) -> str:
        """Run the full orchestration pipeline"""
        plan = await self.plan(task)
        results = await self.delegate(plan)
        return await self.aggregate(results)
```

```python
# Try it!
orchestrator = SimpleOrchestrator(["researcher", "writer", "editor"])
result = await orchestrator.execute("Write a blog post about AI agents")
print(result)
```
