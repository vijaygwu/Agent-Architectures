# Chapter 7: Swarm Pattern Playground

Visualize pheromone-based exploration.

```python
import random
from collections import defaultdict

class PheromoneMap:
    def __init__(self, evaporation_rate=0.1):
        self.pheromones = defaultdict(float)
        self.evaporation_rate = evaporation_rate
    
    def deposit(self, path: str, quality: float):
        self.pheromones[path] += quality
    
    def evaporate(self):
        for path in self.pheromones:
            self.pheromones[path] *= (1 - self.evaporation_rate)
    
    def get_probabilities(self, paths: list) -> dict:
        total = sum(self.pheromones[p] + 0.1 for p in paths)  # +0.1 base
        return {p: (self.pheromones[p] + 0.1) / total for p in paths}
```

```python
# Simulate swarm exploration
pheromone_map = PheromoneMap(evaporation_rate=0.2)
paths = ["path_A", "path_B", "path_C"]
path_quality = {"path_A": 0.3, "path_B": 0.8, "path_C": 0.5}  # B is best

for iteration in range(10):
    probs = pheromone_map.get_probabilities(paths)
    chosen = random.choices(paths, weights=[probs[p] for p in paths])[0]
    pheromone_map.deposit(chosen, path_quality[chosen])
    pheromone_map.evaporate()
    print(f"Iter {iteration}: {dict(pheromone_map.pheromones)}")
```
